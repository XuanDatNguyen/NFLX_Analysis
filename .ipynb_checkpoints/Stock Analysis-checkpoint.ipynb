```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dữ liệu và xem thử dữ liệu
df = pd.read_csv("NFLX.csv")
df.head()
```

```python
#kiểm tra dữ liệu
df.info()
```

```python
#Convert dtype của cột Date từ object -> Datetime
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
#Tách cột year từ cột date
df["Year"] = df["Date"].dt.year
df.head()
```

```python
# Bỏ col Adj Close
df = df.drop(["Adj Close"], axis=1)
df.head()
```

```python
# Gán index cho cột Date
df = df.set_index("Date")
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Close"])
plt.title("Stock Close Price 2002 - 2022", fontsize=24)
plt.xlabel("USD ($)", fontsize=14)
plt.ylabel("year", fontsize=14);
```

```python
#Mức giá đóng cửa cao nhất
df["Close"].max()
```

```python
#Ngày mà có mức giá đóng cửa cao nhất
df["Close"].idxmax()
```

```python
#Mức giá đóng cửa thấp nhất
df["Close"].min()
```

```python
#Ngày mà có mức giá đóng cửa thấp nhất
df["Close"].idxmin()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Volume"])
plt.xlabel('Volume', fontsize = 14)
plt.ylabel('USD ($)', fontsize = 14)
plt.title("Stock Volume From 2002 - 2022", fontsize = 24);
```

```python
# Khối lượng giao dịch lớn nhất
df["Volume"].max()
```

```python
# Ngày có khối lượng giao dịch lớn nhất
df["Volume"].idxmax()
```

```python
# Khối lượng giao dịch nhỏ nhất
df["Volume"].min()
```

```python
df["Volume"].idxmin()
```

```python
# Khối lượng giao dịch trung bình
df["Volume"].mean()
```

### Phân tích kỹ thuật

```python
# Đầu tư ngắn hạn -> lấy dữ liệu từ 2020 trở lại

new_df = df["2020-01-01":]
new_df.head()
```

```python
new_df["Close"].plot(figsize=(10,5), title="Close Price From 2020 To Present");
```

```python
new_df["Volume"].plot(figsize=(10,5), title="Stock Volume From 2020 To Present");
```

#### Đường MA (hay còn gọi là Moving Average) là đường trung bình động, thể hiện biến động, 
#### chỉ báo xu hướng của giá cổ phiếu trong một khoảng thời gian. 
#### Mục đích chính là theo dõi giá đang vận động theo xu hướng tăng hoặc giảm hay không có xu hướng.

```python
# Moving average

new_df["20 Days MA"] = new_df["Close"].rolling(20).mean();
new_df["40 Days MA"] = new_df["Close"].rolling(40).mean();
```

```python
new_df
```

```python
new_df.dropna()
```

```python
new_df[["Close", "20 Days MA", "40 Days MA"]].plot(figsize=(14,7));
```

```python
700/300 - 1
```

```python
new_df["Upper band"] = new_df["20 Days MA"] + 2*new_df["Close"].std()
new_df["Lowwer band"] = new_df["20 Days MA"] - 2*new_df["Close"].std()

new_df.dropna()
```

```python
new_df[["Close", "20 Days MA","Upper band", "Lowwer band"]].plot(figsize=(14,7))
```
