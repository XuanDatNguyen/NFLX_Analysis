```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dữ liệu và xem thử dữ liệu
df = pd.read_csv("NFLX.csv")
df.head()
```

```python
#kiểm tra dữ liệu
df.info()
```

```python
#Convert dtype của cột Date từ object -> Datetime
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
#Tách cột year từ cột date
df["Year"] = df["Date"].dt.year
df.head()
```

```python
# Bỏ col Adj Close
df = df.drop(["Adj Close"], axis=1)
df.head()
```

```python
# Gán index cho cột Date
df = df.set_index("Date")
df.head()
```

### Phân tích kỹ thuật

```python
# Đầu tư ngắn hạn -> lấy dữ liệu từ 2020 trở lại

new_df = df["2020-01-01":]
new_df.head()
```

```python
# Tỷ suất sinh lời từ năm 2020 - 2022
max_price = new_df["Close"].max()
min_price = new_df["Close"].min()

profit_margin = ((max_price / min_price) - 1) * 100
profit_margin

# gửi nh 1 tỷ sau 2 năm áp dụng lãi kép 6,5 %/năm 1,134,224,999
# đầu tư cp nf vốn ban đầu 1,000,000,000 sau 2 năm lãi 1,310,000,000 -> lãi gấp 2.3 lần
```

```python
# Tỷ suất lợi nhuận: so sánh giá trị của giá đóng cửa của ngày n và n-1, profit margin = (Pd/P(d-1) - 1) * 100
new_df["Profit margin"] = new_df["Close"].pct_change()*100
```

```python
new_df["Profit margin"]["2020-01-01":"2020-11-30"].plot(figsize=(10,5), title="Profit Margin");
```

```python
new_df["Close"].plot(figsize=(10,5), title="Close Price From 2020 To Present");
```

```python
new_df["Volume"].plot(figsize=(10,5), title="Stock Volume From 2020 To Present");
```

#### Đường MA (hay còn gọi là Moving Average) là đường trung bình động, thể hiện biến động, 
#### chỉ báo xu hướng của giá cổ phiếu trong một khoảng thời gian. 
#### Mục đích chính là theo dõi giá đang vận động theo xu hướng tăng hoặc giảm hay không có xu hướng.

```python
# Moving average

new_df["20 Days MA"] = new_df["Close"].rolling(20).mean();
new_df["40 Days MA"] = new_df["Close"].rolling(40).mean();
```
